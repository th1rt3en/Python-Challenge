```python
from PIL import Image
```

```python
im = Image.open('cave.jpg')
im
```

```python
w,h = im.size
```

```python
odd = Image.new('RGB', (w/2, h/2))
even = Image.new('RGB', (w/2, h/2))
```

```python
for i in range(w):
    for j in range(h):
        if j%2:2
            odd.putpixel((i/2,j/2), im.load()[i,j])
        else:
            even.putpixel((i/2,j/2), im.load()[i,j])
```
