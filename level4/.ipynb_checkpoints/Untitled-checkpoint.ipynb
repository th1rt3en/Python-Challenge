```python
from urllib import urlopen
import re
```

```python
s = "http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=12345"
while 1:
    t = urlopen(s).read()
    print t
    try:
        s = re.sub(r'\d+',re.findall(r'\d+',t)[0],s)
    except IOError as e:
        print e
```
